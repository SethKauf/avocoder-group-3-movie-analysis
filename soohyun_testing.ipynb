```python
movie_rating_gross = movie_with_rating_with_more_votes.merge(df1, left_on='original_title', right_on='title')
movie_rating_gross.head()
```

Condensed dataframe with title, genre, rating, gross amount

```python
condensed_movie_rating_gross = movie_rating_gross.loc[:, ['primary_title', 'original_title', 'genres', 'averagerating', 'numvotes', 'domestic_gross', 'foreign_gross']]
condensed_movie_rating_gross.head()
```

top 100 movies

```python
top_100 = [title for title in condensed_movie_rating_gross['primary_title']]
top_100
```

Genre

```python
genres = list(condensed_movie_rating_gross['genres'].loc[:100])

top_100_rating_genre_list = []

for genre in genres:
    x = genre.split(',')
    for index in range(len(x)):
        top_100_rating_genre_list.append(x[index])

top_100_rating_genre_dict = {}
for genre in top_100_rating_genre_list:
    if genre in top_100_rating_genre_dict.keys():
        top_100_rating_genre_dict[genre] += 1
    else:
        top_100_rating_genre_dict[genre] = 0
        

top_100_rating_genre_dict = dict(sorted(top_100_rating_genre_dict.items(), key=lambda x : x[1], reverse=True))
top_100_rating_genre_dict
```

See if popular overseas

```python
condensed_movie_rating_gross['foreign_gross'] = condensed_movie_rating_gross['foreign_gross'].str.replace(',', '').astype(float)
```

```python
condensed_movie_rating_gross.sort_values(by=['foreign_gross'], ascending=False).head()
```

if making films, what genre

```python
budget_movie = df11.merge(condensed_movie_rating_gross, left_on='movie', right_on='primary_title')
budget_movie.head()
```

```python
budget_movie['production_budget'] = budget_movie['production_budget'].str.replace(',', '').str[1:].astype(float)
budget_movie['domestic_gross_x'] = budget_movie['domestic_gross_x'].str.replace(',', '').str[1:].astype(float)
budget_movie['worldwide_gross'] = budget_movie['worldwide_gross'].str.replace(',', '').str[1:].astype(float)
budget_movie.info()
```

```python
budget_movie.rename(columns={'averagerating':'rating'}, inplace=True)
budget_movie.head()
```

```python
budget_movie['profit'] = budget_movie['worldwide_gross'] - budget_movie['production_budget']
```

```python
budget_movie.head()
```

```python
budget_movie.sort_values(by=['profit'], ascending=False, inplace=True)
```

```python
genres = list(budget_movie['genres'].loc[:100])

top_100_profit_genre_list = []

for genre in genres:
    x = genre.split(',')
    for index in range(len(x)):
        top_100_profit_genre_list.append(x[index])

top_100_profit_genre_dict = {}
for genre in top_100_profit_genre_list:
    if genre in top_100_profit_genre_dict.keys():
        top_100_profit_genre_dict[genre] += 1
    else:
        top_100_profit_genre_dict[genre] = 0
        

top_100_profit_genre_dict = dict(sorted(top_100_profit_genre_dict.items(), key=lambda x : x[1], reverse=True))
top_100_profit_genre_dict
```
